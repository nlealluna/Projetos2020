#####Regressão

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

```python
data_cleaned.head()
```

```python
data_with_dummies =pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
```

```python
data_preprocessed.head()
```

Modelo de regressão linear

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs.head()
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

###O GRAFICO RESIDUAL MOSTRA: DISTRIBUIÇÃO NORMAL; MEDIA EM TORNO DE 0 (ESSAS DUAS COISAS DIZEM QUE O MODELO É BOM); CONTUDO APRESENTA UMA HETEROGENOIDADE NAS CAUDAS


```python
reg.score(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

### Perfomance do modelo


```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_train_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

### Concentração dos dados no meio!

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residual']= df_pf['Target'] - df_pf['Prediction']

df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
